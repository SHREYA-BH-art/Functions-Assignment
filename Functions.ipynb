# **Theory Questions**

 1. What is the difference between a function and a method in Python?
 A. A **function** is a block of code that performs a specific task, defined using `def`. A **method** is a function that is associated with an object and is called on that object.

Example:
- Function: `def add(a, b): return a + b`
- Method: `"hello".upper()`

2. Explain the concept of function arguments and parameters in Python.
A. In Python, parameters are the variables defined in the function signature that specify what values the function expects. Arguments are the actual values passed to the function when it is called. When defining a function, you specify the parameters, and when calling the function, you provide the corresponding arguments.

For example, in a function that greets someone, the parameters could be name and age, while the arguments passed during the function call could be "Alice" and 30. Parameters act as placeholders for the values, while arguments are the real values provided during the function execution.

3. What are the different ways to define and call a function in Python?
A. In Python, functions can be defined using the `def` keyword, followed by the function name and parameters in parentheses. Functions can also be defined using lambda expressions for simple, anonymous functions.

To call a function, you use its name followed by parentheses, optionally passing arguments inside them.

For example, a function is defined with def greet(name): and called as greet("Alice"). Functions can also be called by passing arguments by position or by keyword. Lambda functions are invoked like regular functions but are more concise and typically used for short operations.

4. What is the purpose of the retur statement in a Python function?
A. The return statement in Python is used to send a result back from a function to the caller. It allows the function to output a value, which can then be used in other parts of the program. Without a return statement, a function executes but does not provide any output. The return statement can also terminate the function's execution early if needed. For example, a function that calculates the sum of two numbers would use return to send the sum back to where the function was called.

5. What are iterators in Python and how do they differ from iterables?
 A. In Python, an iterator is an object that allows you to traverse through all the elements of a collection, such as a list or tuple, one at a time. An iterator implements two methods: __iter__() and __next__(). An iterable, on the other hand, is any object capable of returning an iterator, meaning it can be looped over (e.g., lists, tuples, dictionaries). The main difference is that iterables need to be converted into iterators using iter(), while iterators directly provide elements through next(). For example, a list is iterable but not an iterator.

6. Explain the concept of generators in Python and how they are defined
A. Generators in Python are special types of iterators that allow lazy evaluation, meaning they generate values one at a time as they are needed, instead of computing all values at once and storing them in memory. This makes them memory-efficient for working with large datasets. A generator is defined using a function with the yield keyword, which pauses the function's execution and returns a value. When the generator is called again, it resumes from where it left off. Generators are commonly used in scenarios where processing large amounts of data or performing operations on-the-fly is required.

7. What are the advantages of using generators over regular functions?
A. Generators offer several advantages over regular functions. First, they are memory-efficient because they yield one item at a time, rather than storing all results in memory. This is particularly useful for working with large datasets. Second, generators are faster for iterating over data as they do not require creating and storing an entire collection in memory. Additionally, they can represent infinite sequences, which regular functions cannot easily do. Overall, generators provide a more efficient way to handle large or continuous streams of data.

8. What is a lambda function in Python and when is it typically used?
A. A lambda function in Python is a small, anonymous function defined using the lambda keyword, followed by parameters, a colon, and an expression. Unlike regular functions defined with def, lambda functions can have only one expression and do not require a return statement. They are typically used for short-term, throwaway operations where defining a full function would be overkill. Common use cases include operations like sorting, filtering, or mapping data in a concise manner. For example, a lambda function can be used to perform quick calculations or pass as an argument to higher-order functions like map() or filter().

9. Explain the purpose and usage of the `map()` function in Python.
A. The map() function in Python is used to apply a given function to all items in an iterable (e.g., list, tuple) and return a map object (an iterator) containing the results. It allows you to process and transform the elements of an iterable efficiently. The map() function takes two arguments: the function to apply and the iterable. It's commonly used for tasks like transforming data, such as applying mathematical operations or converting values in a list. The result can be converted to a list or another iterable type for further use.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
A. The `map()`, `reduce()`, and `filter()` functions in Python are used for functional programming but serve different purposes:

- map(): Applies a given function to all items in an iterable and returns a new iterable with the results. It's used for transforming data, such as applying a function to each element of a list.
  
- reduce(): From the `functools` module, it applies a binary function cumulatively to the items of an iterable, reducing them to a single value. It's often used for accumulation or aggregation operations.
  
- filter(): Filters elements from an iterable based on a function that returns a Boolean value, returning only the items that meet the condition.

```python
1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.
A. def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
```

```python
4.  Write a Python function that checks if a given number is prime or not from 1 to 200.
A. def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

```python
5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.
A. class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num)
```

```python
6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
A. def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

```python
7.  Implement a generator function that reads a file line by line and yields each line as a string.
A. def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_file_line_by_line('example.txt'):
    print(line)
```

```python
8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
A. my_list = [(1, 3), (2, 1), (4, 2)]
sorted_list = sorted(my_list, key=lambda x: x[1])
print(sorted_list)
```

```python
9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
A. celsius = [0, 20, 30, 40, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
A. input_string = "Hello World"
vowels = "aeiouAEIOU"
filtered_string = ''.join(filter(lambda x: x not in vowels, input_string))
print(filtered_string)
```
